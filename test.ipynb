### 計算每個月應還得本金+利息包含本利和跟餘額

```python
interest = round(loan_amout*(year_rate/100/12))
```

```python
loan_amout=100000
year_rate=3.5
year_priod=7


amount=loan_amout
month=year_priod*12
month_rate=year_rate/100/12
month_pay=int(loan_amout/(year_priod*12))

for i in range(month):
    
    interest = round(amount*month_rate)
    amount-=month_pay
    if i==month-1:
        print(f'期別:{i+1}本金:{month_pay+amount}利息:{interest}本利和:{month_pay+interest+amount}餘額:{0}')
    else:
        print(f'期別:{i+1}本金:{month_pay}利息:{interest}本利和:{month_pay+interest}餘額:{amount}')
    
```
